```python
ll =[]
cc = []
for a in ['accepted','partial','excluded']:
    for b in range(0,24,1):
        ll.append((a,b))
```

```python
plt.figure(figsize =(24,8), dpi=150)
nrv['hour'] = nrv.time.apply(lambda x: int(x.split(":")[0]))
plot = nrv.groupby('tweetType').hour.value_counts().reindex(ll).plot(kind='bar')
```

```python
plt.figure(figsize =(24,8), dpi=150)
big['hour'] = nrv.time.apply(lambda x: int(x.split(":")[0]))
plot = big.groupby('tweetType').hour.value_counts().reindex(ll).plot(kind='bar')
```

```python
words_nrv = []
for x in nrv.text:
    y = x.split(" ")
    words_nrv.extend(y)
wn = pd.DataFrame(words_nrv,columns=['words'])
wn.words.value_counts()[:20].plot(kind='bar')
```

```python
words_big = []
for x in big.text:
    y = x.split(" ")
    words_big.extend(y)
wb = pd.DataFrame(words_big,columns=['words'])
wb.words.value_counts()[:20].plot(kind='bar')
```
