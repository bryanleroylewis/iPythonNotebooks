```python
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_grid_id(data,box,step_size):
    """Define an id for a grid cell, takes a box with min,max coords for 2d and stepsize.  
    Starts in bottom left (min,min) and going right to left (x1,x2,x3..xn, up to next y row, x1,x2..xn, etc.)
    Counts the cells"""
    
    x_size = box["x_max"] - box["x_min"]
    y_size = box["y_max"] - box["y_min"]
    
    x_dim = math.ceil(x_size / step_size)
    y_dim = math.ceil(y_size / step_size)
    
    x_off = data['longitude'] - box["x_min"]
    y_off = data['latitude'] - box["y_min"]
    
    cell_id = math.floor(y_off / step_size) * x_dim + math.floor(x_off / step_size)
    
    return cell_id
```

```python
def grid_id_to_lat(data,box,step_size):
    x_size = box["x_max"] - box["x_min"]
    y_size = box["y_max"] - box["y_min"]
    
    x_dim = math.ceil(x_size / step_size)
    y_dim = math.ceil(y_size / step_size)
    
    row = int( math.floor(data['grid_id'] / x_dim) + 1)
    col = int(data['grid_id'] - (row -1) * x_dim)

    lat = box['y_min'] + row*step_size
    lon = box['x_min'] + col*step_size

    ##print str(row) + " " + str(col)
    
    return lat
```

```python
def grid_id_to_lon(data,box,step_size):
    x_size = box["x_max"] - box["x_min"]
    y_size = box["y_max"] - box["y_min"]
    
    x_dim = math.ceil(x_size / step_size)
    y_dim = math.ceil(y_size / step_size)
    
    row = int( math.floor(data['grid_id'] / x_dim) + 1)
    col = int(data['grid_id'] - (row -1) * x_dim)

    lat = box['y_min'] + row*step_size
    lon = box['x_min'] + col*step_size

    ##print str(row) + " " + str(col)
    
    return lon
```

```python
def distance_to_point(data,lat,lon):
    ''' Specific to DC area for meter conversion '''
    lat_dist = data['lat'] - lat
    lon_dist = data['lon'] - lon
    
    lat_km = lat_dist * 110.996
    lon_km = lon_dist * 87.832
    
    dist = math.sqrt(lat_km*lat_km + lon_km*lon_km)
    
    return dist    
```

```python
vul = pd.read_csv("WashDC_vul_latlong.csv",names=('id','vul','longitude','latitude'),skiprows=1)
```

```python
box = {}
box["y_min"] = vul.latitude.min()
box["y_max"] = vul.latitude.max()
box["x_min"] = vul.longitude.min()
box["x_max"] = vul.longitude.max()
box
```

```python
vul['grid_id'] = vul.apply(get_grid_id,axis=1,args=(box,0.0001))
```

```python
mean_vuls = vul.groupby('grid_id').vul.mean()
```

```python
median_vuls = vul.groupby('grid_id').vul.median()
```
